```python
import pandas as pd
import numpy as np
from sklearn import linear_model 
```

```python
df=pd.read_csv(r"C:\Users\tirth\Desktop\CSV files\ml2.csv")
df
```

```python
import math
median_bedrooms=math.floor(df.bedrooms.median()) #median of bedrooms
median_bedrooms

```

```python
df.bedrooms=df.bedrooms.fillna(median_bedrooms) #filling the missing values with median of bedrooms
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price) 
#df.price is the dependent variable and df[['area','bedrooms','age']] is the independent variable
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[4000,5,20]])
```

```python
# graph
import matplotlib.pyplot as plt
plt.plot(df.area,df.price)
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```
